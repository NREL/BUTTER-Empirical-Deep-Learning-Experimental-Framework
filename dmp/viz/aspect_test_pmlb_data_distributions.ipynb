```python
datasets = ['529_pollen',
        'sleep',
        'adult',
        '537_houses',
        'nursery',
        '201_pol',
        'mnist',
        'connect_4',
        'wine_quality_white']

import matplotlib.pyplot as plt
from dmp.data.pmlb import pmlb_loader
from dmp.data.pmlb.pmlb_loader import load_dataset
```

# Distribution of Output

```python
pmlb_index = pmlb_loader.load_dataset_index()

fig, axs = plt.subplots(2,len(datasets)//2+1, figsize=(20,10))
axs = axs.flatten()

for i, dataset in enumerate(datasets):
    ds, inputs, outputs = load_dataset(pmlb_index, dataset)
    if outputs.shape[1] > 1:
        outputs = [outputs[j].argmax() for j in range(outputs.shape[0])]
    axs[i].scatter(list(range(len(outputs))), outputs, marker=".")
    axs[i].set_title(dataset)

plt.show()
    

```

# Distribution of Input[0]

```python
pmlb_index = pmlb_loader.load_dataset_index()

fig, axs = plt.subplots(2,len(datasets)//2+1, figsize=(20,10))
axs = axs.flatten()

for i, dataset in enumerate(datasets):
    ds, inputs, outputs = load_dataset(pmlb_index, dataset)
    axs[i].scatter(list(range(len(inputs[:,0]))), inputs[:,0], marker=".")
    axs[i].set_title(dataset)

plt.show()
```
